# What is ggsolver

ggsolver is a Python package for constructing, manipulating, and solving various classes of games on graphs.


**Important Links:**
* Documentation: [https://akulkarni.me/ggsolver/](https://akulkarni.me/ggsolver/)
* GitHub: [https://github.com/abhibp1993/ggsolver](https://github.com/abhibp1993/ggsolver)


# Important parts of ggsolver

**Important Modules:**
* `graphs.py`: Basic data structure
* `models.py`: Basic models

**Important Packages:**
* `logic`: Provides LTL and ScLTL classes.
* `dtptb`: Modeling and solving deterministic two-player turn-based games.
* `mdp`: Modeling and solving games on Markov decision processes.



# How to use ggsolver in my project?

**Model:**
1. What type of game do I want to solve?
2. Does there exist a package that provides the model?
3. If not, what’s the best model that I can derive from?

**Languages:**
1. What language will I express the objectives?
2. Does the logic exist in “logic” package?
3. If not, talk to professor or TAs.

**Algorithms/Solver:**
1. What algorithm do I want to use to solve the game?
2. Does there exist a package that implements the algorithm for that model?

**Visualize and Analyze:**
1. How will I simulate my game?
2. Gridworld simulators: see “gridworld” package (talk to Abhishek)
3. Different simulators: Input serialized graph generated by ggsolver into your simulators.
